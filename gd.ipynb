```python
import matplotlib.pyplot as plt
from ipywidgets import interact
import numpy as np
import random
%matplotlib inline
```

```python
# 列出方程式
方程式:
L(x)=x4−2x3+1x2+x

微分後的方程式:
L2(x)=12x3−6x2+2x+1
```

```python
def L(x):
    return x**4 - 5*x**3 + 1*x
def L2(x):
    return 4*x**3 - 15*x**2  + 1
```

```python
#畫圖
x = np.linspace(-4,5, 1000)
plt.plot(x,L(x))
```

```python
#起始位置
start = random.uniform(-4,5)
print("start:",start)
#畫結果
lr=0.04
plt.plot(x,L(x))
#最多跌代100次
for i in range(100):
    if L2(start) == 0 :
        break
    else:
        start = start - lr*L2(start)
        plt.scatter(start,L(start))
```
